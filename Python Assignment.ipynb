```python
#Create 2 list with some brand names
l1 = ["LOUIS VUITTON" , "GUCCI" , "HERMES" , "PRADA" , "CHANEL"]
l2 = ["NIKE" , "LOUIS VUITTON" , "HERMES" , "GUCCI" , "ZALANDO"]
```

```python
#Sort the list
print(sorted(l1))
print(sorted(l2))
```

```python
#List of overlapping brand names in both lists
common_list=[]
for item in l1:
    if item in l2:
        common_list += [item]
print(common_list)
```

```python
#List of overlapping brand names in both lists
print(list(set(l1) & set(l2)))
```
